```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data resembling experimental values
np.random.seed(42)
curing_days = np.logspace(-3, 3, 30)
true_curve = 60 / (1 + (curing_days / 0.5)**2)
experimental_values = true_curve + np.random.normal(0, 2, len(true_curve))

# Plot
plt.figure(figsize=(8, 6))

# Experimental values (scatter plot)
plt.scatter(curing_days, experimental_values, color='black', label='Experimental value', facecolors='none')

# Fitted curve (logarithmic scale)
plt.plot(curing_days, true_curve, color='black', label='Fitted value')

# Formatting the plot
plt.xscale('log')
plt.ylim(0, 70)
plt.xlabel('Curing day (days)')
plt.ylabel('Weight fraction (g/100g of binder)')
plt.title('Weight Fraction vs Curing Day')
plt.legend()
plt.grid(which="both", linestyle="--", linewidth=0.5)

plt.show()
```

```python
import numpy as np
import pandas as pd

# Generate random data resembling experimental values
np.random.seed(42)
curing_days = np.logspace(-3, 3, 30)
true_curve = 60 / (1 + (curing_days / 0.5)**2)
experimental_values = true_curve + np.random.normal(0, 2, len(true_curve))

# Create a DataFrame
data = {
    "Curing_Days (days)": curing_days,
    "True_Curve (g/100g)": true_curve,
    "Experimental_Values (g/100g)": experimental_values
}
df = pd.DataFrame(data)

# Save to a CSV file
file_name = "curing_data.csv"
df.to_csv(file_name, index=False)

print(f"Data saved to {file_name}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define Option 3 function
def option3_fitting_function(x):
    return 41.57 * np.exp(-2.00 * x) + 16.85

# Generate smooth data for plotting
x_data = np.linspace(0, 30, 500)  # Smooth range for plotting
y_data = option3_fitting_function(x_data)

# Original data
X = np.array([0, 1, 3, 7, 28])
Y = np.array([58.42, 19.30, 18.10, 16.50, 13.50])

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Original Data')
plt.plot(x_data, y_data, color='red', label='Option 3: $y = 41.57e^{-2.00x} + 16.85$')

# Formatting
plt.xlabel('X (Days)')
plt.ylabel('Y (Alite)')
plt.title('Curve Fitting with Option 3')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Original data
X = np.array([0, 1, 3, 7, 28])
Y = np.array([58.42, 19.30, 18.10, 16.50, 13.50])

# Define the fitting function with 'c' fixed at 13.50 to ensure the curve passes through the last data point
def fitting_function(x, a, b):
    return a * np.exp(-b * x) + 13.50  # c is fixed at 13.50

# Initial guess for parameters 'a' and 'b'
initial_guess = [41.57, 2.00]

# Perform curve fitting with 'c' fixed
params, _ = curve_fit(fitting_function, X, Y, p0=initial_guess)

# Generate smooth data for plotting
x_smooth = np.linspace(min(X), max(X), 500)
y_smooth = fitting_function(x_smooth, *params)

# Calculate fitted values at the original X points
fitted_Y = fitting_function(X, *params)

# Plot the original data and the fitted curve
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Original Data', zorder=5)
plt.plot(x_smooth, y_smooth, color='red', label=f'Fitted Curve: $y = {params[0]:.2f}e^{{-{params[1]:.2f}x}} + 13.50$')

# Formatting
plt.xlabel('X (Days)')
plt.ylabel('Y (Alite)')
plt.title('Adjusted Curve Fitting to Match Last Data Point')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.show()

# Print the adjusted parameters
print(f"Adjusted Parameters with c fixed at 13.50:\na = {params[0]:.2f}\nb = {params[1]:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Data
X = np.array([0, 1, 3, 7, 28])
CO2 = np.array([0.0, 5.0, 12.0, 18.0, 20.0])

# Define the fitting function (e.g., logistic or exponential)
def fitting_function(x, a, b, c):
    return a * (1 - np.exp(-b * x)) + c  # Example: Exponential growth with saturation

# Perform curve fitting
initial_guess = [20, 0.1, 0]  # Initial guess for parameters a, b, c
params, covariance = curve_fit(fitting_function, X, CO2, p0=initial_guess)

# Generate smooth data for plotting
x_smooth = np.linspace(min(X), max(X), 500)
y_smooth = fitting_function(x_smooth, *params)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(X, CO2, color='blue', label='Original Data', zorder=5)
plt.plot(x_smooth, y_smooth, color='red', label=f'Fitted Curve: $y = {params[0]:.2f}(1 - e^{{-{params[1]:.2f}x}}) + {params[2]:.2f}$')

# Formatting
plt.xlabel('X (Days)')
plt.ylabel('CO₂ (%)')
plt.title('Curve Fitting for CO₂ vs Days')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.show()

# Display fitted parameters
print(f"Fitted Parameters:\na = {params[0]:.2f}\nb = {params[1]:.2f}\nc = {params[2]:.2f}")
```
